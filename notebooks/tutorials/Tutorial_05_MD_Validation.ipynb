Run short OpenMM simulations on top candidates to prioritize stable buried pockets.


## Workflow
- Load top candidates from screening
- Configure MD settings
- Run validation pipeline
- Classify pockets by stability behavior


```python
from cryptic_ip.validation.md_validation import OpenMMMDValidationPipeline, MDSimulationConfig

config = MDSimulationConfig(simulation_time_ns=5.0, temperature_kelvin=300)
pipeline = OpenMMMDValidationPipeline(output_dir='results/md_tutorial', simulation_config=config)
```

```python
report = pipeline.validate_top_candidates('results/combined_hits.csv', top_n=10)
report.head()
```
