Screen a full proteome directory with parallel workers and checkpoint-friendly outputs.


## Workflow
- Prepare AlphaFold structure directory
- Build proteome catalog
- Run parallel screen
- Summarize top hits


```python
from cryptic_ip.database import ProteomeManager
from cryptic_ip.database.batch_processing import ParallelProcessor

manager = ProteomeManager('data/structures/yeast')
catalog = manager.build_catalog()
catalog.head()
```

```python
processor = ParallelProcessor(n_jobs=8)
results = processor.process_catalog(catalog, score_threshold=0.6, use_ml_model=True)
results.head()
```

```python
results.to_csv('results/tutorial03_yeast_hits.csv', index=False)
```
