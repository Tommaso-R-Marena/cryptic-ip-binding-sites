```python
from cryptic_ip.analysis import ProteinAnalyzer
from cryptic_ip.database.pdb_client import PDBClient

pdb = PDBClient()
adar2_path = pdb.download_structure('1ZY7', output_dir='data/tutorial')
adar2_path
```

```python
analyzer = ProteinAnalyzer(adar2_path)
scored = analyzer.score_all_pockets()
scored.sort_values('composite_score', ascending=False).head(10)
```
