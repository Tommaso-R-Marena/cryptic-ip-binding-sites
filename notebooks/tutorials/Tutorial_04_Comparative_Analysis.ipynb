Compare organism-level hit rates, run statistics, and generate publication-ready figures.


## Workflow
- Load per-organism screening outputs
- Run comparative statistics
- Generate confidence intervals and effect sizes
- Create summary plots


```python
import pandas as pd
from cryptic_ip.analysis.comparative_analysis import ComparativeIPAnalysis

yeast = pd.read_csv('results/yeast_hits.csv')
human = pd.read_csv('results/human_hits.csv')
dicty = pd.read_csv('results/dicty_hits.csv')
```

```python
comp = ComparativeIPAnalysis()
result = comp.compare({'S_cerevisiae': yeast, 'H_sapiens': human, 'D_discoideum': dicty})
result.summary
```

```python
result.figures['hit_rates'].savefig('results/tutorial04_hit_rates.png', dpi=300)
```
